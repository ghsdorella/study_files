```
import pandas as pd
import matplotlib.pyplot as plt

%precision %.2f
pd.options.display.float_format = '{:,.2f}'.format
```

```
url = 'https://raw.githubusercontent.com/guilhermesilveira/enem-2018/master/MICRODADOS_ENEM_2018_SAMPLE_43278.csv'
dados = pd.read_csv(url)
dados.head()
```

```
dados.columns
```

```
dados.describe()
```

```
colunas_notas = ['NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC', 'NU_NOTA_MT', 'NU_NOTA_REDACAO']
dados_notas = dados[colunas_notas].dropna()
dados_notas.columns = ['ciencias_naturais', 'ciencias_humanas', 'linguagem_codigo', 'matematica', 'redacao']
dados_notas.head()
```

```
corr = dados_notas.corr()
corr
```

```
from string import ascii_letters
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=np.bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))


# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap="YlGnBu", square=True, linewidths=.5, vmin=0, annot=True)
```

```
sns.pairplot(dados_notas, kind='reg')
```

Podemos usar também o método SimpleImputer() para imputar valores de medianas nessas colunas, como o exemplo a baixo


```
#Criando um novo dataset
dados_com_mediana = pd.read_csv(url)
```

```
dados_notas3 = dados_com_mediana[colunas_notas]
dados_notas3.columns = ['ciencias_naturais', 'ciencias_humanas', 'linguagem_codigo', 'matematica', 'redacao']
dados_notas3.head()
```

```
from sklearn.impute import SimpleImputer

my_imputer = SimpleImputer(strategy='median')
dados_notas_mediana = pd.DataFrame(my_imputer.fit_transform(dados_notas3))
```

```
dados_notas_mediana.columns = dados_notas3.columns
```

```
corr3 = dados_notas_mediana.corr()
corr3
```

```
# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr3, dtype=np.bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))


# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr3, mask=mask, cmap="YlGnBu", square=True, linewidths=.5, vmin=0, annot=True)
```

```
def acertos_cn(a):
  gabarito =  dados['TX_GABARITO_CN'][a]
  respostas = dados['TX_RESPOSTAS_CN'][a]
  x=0
  if type(gabarito) == float or type(respostas) == float:
    x = 0
  else:
    for i, j in zip(gabarito, respostas):
      if i == j:
        x += 1
  return x

dados['acertos_ciencias_naturais'] = list(map(acertos_cn, range(0,len(dados))))
```

```
def acertos_ch(a):
  gabarito =  dados['TX_GABARITO_CH'][a]
  respostas = dados['TX_RESPOSTAS_CH'][a]
  x=0
  if type(gabarito) == float or type(respostas) == float:
    x = 0
  else:
    for i, j in zip(gabarito, respostas):
      if i == j:
        x += 1
  return x

dados['acertos_ciencias_humanas'] = list(map(acertos_ch, range(0,len(dados))))
```

```
def acertos_mt(a):
  gabarito =  dados['TX_GABARITO_MT'][a]
  respostas = dados['TX_RESPOSTAS_MT'][a]
  x=0
  if type(gabarito) == float or type(respostas) == float:
    x = 0
  else:
    for i, j in zip(gabarito, respostas):
      if i == j:
        x += 1
  return x

dados['acertos_matematica'] = list(map(acertos_mt, range(0,len(dados))))
```
